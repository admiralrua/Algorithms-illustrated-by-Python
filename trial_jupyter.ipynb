```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

plt.subplot(1, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('First plot')
plt.ylabel('Damped oscillation')
plt.xlabel('Time (s)')
plt.show()
```
